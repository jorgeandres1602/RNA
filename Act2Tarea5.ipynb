```python
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
import random
import matplotlib.pyplot as plt


model = Sequential()
model.add(Dense(32, input_dim=1, activation='relu'))
model.add(Dense(1, activation='linear'))
```

```python
x_train = np.linspace(-1, 1, num=50)
y_train = 3 * np.sin(np.pi * x_train)


```

```python
model.compile(loss='mean_squared_error', optimizer='adam')
```

```python
model.fit(x_train, y_train, epochs=1000, batch_size=1)
```

```python
y_pred = model.predict(x_train)
```

```python
plt.plot(x_train, y_pred, label="aprox")
plt.plot(x_train, y_train,label="exact")
plt.legend()
plt.show()
```

```python
x_train2 = np.linspace(-1, 1, num=50)
y_train2 = 1 + 2 * x_train + 4 * (x_train)**3
```

```python
model.fit(x_train2, y_train2, epochs=1000, batch_size=1)
```

```python
y_pred2 = model.predict(x_train2)
```

```python
plt.plot(x_train2, y_pred2, label="aprox")
plt.plot(x_train2, y_train2,label="exact")
plt.legend()
plt.show()
```
