```python
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
import random
import matplotlib.pyplot as plt


model = Sequential()
model.add(Dense(32, input_dim=1, activation='relu'))
model.add(Dense(1, activation='linear'))
```

```python
a=[random.randint(0,1) for i in range(4)]
x_train = np.linspace(-1, 1, num=50)
y_train = a[0] + a[1] * x_train + a[2] * (x_train)**2 + a[3] * (x_train)**3
```

```python

model.compile(loss='mean_squared_error', optimizer='adam')
```

```python
model.fit(x_train, y_train, epochs=1000, batch_size=1)
```

```python
y_pred = model.predict(x_train)
```

```python
plt.plot(x_train, y_pred, label="aprox")
plt.plot(x_train, y_train,label="exact")
plt.legend()
plt.show()
```

```python
y_train2=np.cos(2*x_train)
```

```python
model.fit(x_train, y_train2, epochs=1000, batch_size=1)
```

```python
y_pred2 = model.predict(x_train)
```

```python
plt.plot(x_train, y_pred2, label="aprox")
plt.plot(x_train, y_train2,label="exact")
plt.legend()
plt.show()
```
